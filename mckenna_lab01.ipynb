## Lab 01 - Machine Learning

```python
import pandas as pd
import numpy as np
```

1. Create a tiny dataset (8-12 rows and 2-3 columns max) using a Pandas DataFrame.

```python
data = {
    'Coffee Shop': ['Ren', 'The Well', 'Falls Perk', 'Shortsleeves', 'Brightside', 'Invited', 'Cartel Roasting Co', 'Bovacanti'],
    'City': ['Tucson', 'Fishers', 'Pendleton', 'Swannanoa', 'Scottsdale', 'Pendleton', 'Tucson', 'Carmel'],
    'Favorite Drink': ['Blueberry Latte', 'Charcol Latte', 'Iced Lavander Latte', 'Old Fashioned Coffee', 'Cookie Butter Latte', 'Rasberry Latte', 'Iced Latte', 'Iced Latte'],
    'Miles Away': [11, 1, 1722, 1737, 1927, 110, 11, 1720]
}
df = pd.DataFrame(data)
print(df)
```

2. Convert one of the columns of the DataFrame to numpy and calculate the mean and standard deviation of that variable.

```python
miles_away = df['Miles Away'].to_numpy()

mean_miles = np.mean(miles_away)
print(f"Mean miles away: {mean_miles}")

miles_std = np.std(miles_away)
print(f"Standard deviation of miles away: {miles_std}")
```
